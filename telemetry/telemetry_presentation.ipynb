```python
laps = race.laps.pick_driver('VER')

telemetry_list = []

for index, lap in laps.iterlaps():
    telemetry = lap.get_telemetry()
    telemetry['LapNumber'] = lap['LapNumber']
    
    telemetry_list.append(telemetry)
    
telemetry = pd.concat(telemetry_list)
```
